```julia
using Plots
```

```julia
T=10
λ=1/2
t() = -1/λ*log(rand())
n=Int(1e3)
Δ=T/n
;
```

# Zad 1

```julia
a=[t()]
```

```julia
while a[end]<T
    push!(a,t()+a[end])
end
```

```julia
a
```

```julia
plot(a,0:length(a)-1, linetype=:steppre, leg=false)
```

```julia
b=zeros(n);
```

```julia
j=1
for i in 1:n
    while i*Δ>=a[j]
        j+=1
    end
    b[i]=j
end
    
b=b.-1;        
    
```

```julia
scatter!(0:Δ:T-Δ,b)
```

```julia
scatter!(xlim=(0,10))
```
